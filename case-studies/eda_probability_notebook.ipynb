```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
# Bernoulli
p = 0.3
n = 1000

bernoulli_sample = np.random.binomial(1, p,size=n)

plt.hist(bernoulli_sample, bins=2, edgecolor = 'black')
plt.title('Loi de Bernouilli (p=0.3)')
plt.xticks([0, 1])
plt.xlabel('Valeur')
plt.ylabel('Frequence')
plt.show()
```

```python
# Binomiale

n_essais = 10
p = 0.5
bino_sample = np.random.binomial(n_essais, p, size=1000)

plt.hist(bino_sample, bins=range(0, n_essais+2), edgecolor='black', align='left')
plt.title('Loi binomiale (p=0.5)')
plt.xlabel("Nombre de succes")
plt.ylabel('Frequence')
plt.show()
```

```python
# geometrique

p = 0.3
n = 1000

geo_sample = np.random.geometric(p, size=n) - 1

plt.hist(geo_sample, bins=range(0, max(geo_sample)+1), edgecolor='black', density=True, align='left')
plt.title('Loi geometrique (p=0.3)')
plt.xlabel("Nombre d'echecs")
plt.ylabel('Probabilite')
plt.grid(True)
plt.show()
```

```python
# Poisson

lambda_poisson = 4
poisson_sample = np.random.poisson(lambda_poisson, size=1000)

plt.hist(poisson_sample, bins=range(0, max(poisson_sample)+1), edgecolor='black', align='left')
plt.title('Loi de Poisson (lambda = 4)')
plt.xlabel("Nbre d'evenements")
plt.ylabel('Frequence')
plt.show()
```

```python
# exponentielle

lambda_exp = 1.5
exp_sample = np.random.exponential(1/lambda_exp, size=1000)

plt.hist(exp_sample, bins=30, density=True, edgecolor='black')
plt.title('Loi exponentielle (lambda = 1.5)')
plt.xlabel('Temps entre evenements')
plt.ylabel("Densite")
plt.show()
```

```python
# Normale

mu = 0
sigma = 0.5
normal_sample = np.random.normal(mu, sigma, size=1000)

plt.hist(normal_sample, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black')
plt.title("Loi Normale (mu = 0 et sigma = 1)")
plt.xlabel('Valeur')
plt.ylabel('Densite')
plt.show()
```
