# Exploratory Data Analysis (EDA) on the Chinook Database

```python
# Import necessary libraries
import sqlite3
import pandas as pd
```

```python
# Connect to the SQLite database
conn = sqlite3.connect('../data/Chinook_Sqlite.sqlite')
```

```python
# List all tables in the database
tables = pd.read_sql_query("SELECT name FROM sqlite_master WHERE type='table';", conn)

tables
```

```python
# Load the 'customers' table into a DataFrame
customers_df = pd.read_sql_query("SELECT * FROM customer", conn)

customers_df.head(10)
```

```python
# Load the 'invoices' table into a DataFrame
invoices_df = pd.read_sql_query("SELECT * FROM invoice", conn)

invoices_df.head(10)
```

```python
# load the customers with invoices
customers_invoices_df = pd.read_sql_query("""
SELET *
FROM customer as c
JOIN invoice as i ON c.CustomerId = i.CustomerId
""", conn)
```
